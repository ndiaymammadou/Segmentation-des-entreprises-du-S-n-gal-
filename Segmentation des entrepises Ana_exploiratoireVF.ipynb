### I.1.4 Detection et corrections des outliers

Pour détecter et supprimer des valeurs aberrantes dans notre base de données, on définie une fonction **outliers()**. Pour chaque variable numérique (type int32 ou float64), elle calcule les quantiles aux 0.05 (5%) et 0.95 (95%) par branches d'activités et remplace par NAles valeurs qui sortent de l'intervalle.

```python
def outliers(df, group_col='BRANCHE', numeric_cols=None, lower_q=0.05, upper_q=0.95, inplace=True):
    """
    Applique la suppression des outliers par groupe BRANCHE sur les colonnes numériques.
    - df: DataFrame d'entrée
    - group_col: nom de la colonne de regroupement
    - numeric_cols: liste des colonnes numériques à traiter; si None, auto-détection
    - lower_q, upper_q: quantiles pour délimiter les outliers
    - inplace: s'il faut modifier le DataFrame d'entrée ou retourner une copie
    """
    if numeric_cols is None:
        numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
        # Optionnel: retirer la colonne de regroupement si elle est numérique
        if group_col in numeric_cols:
            numeric_cols.remove(group_col)

    # Fonction appliquée à chaque groupe
    def process_group(subdf):
        sub = subdf.copy()
        for col in numeric_cols:
            if col in sub.columns:
                ql = sub[col].quantile(lower_q)
                qu = sub[col].quantile(upper_q)
                mask = (sub[col] < ql) | (sub[col] > qu)
                sub.loc[mask, col] = np.nan  # ou remplacer par une autre valeur, selon votre choix
        return sub

    # Appliquer par groupe
    if group_col in df.columns:
        result = df.groupby(group_col, group_keys=False).apply(process_group)
    else:
        # Pas de regroupement par BRANCHE: appliquer directement
        result = process_group(df)

    return result if not inplace else df.update(result) or df

```

```python
data = outliers(data, group_col='BRANCHE', numeric_cols=None, lower_q=0.05, upper_q=0.95, inplace=False)
```

### I.1.5 Création des variables

* **Solvabilité**

Le ratio de solvabilité obtenu par le rapport **« capitaux propres sur dette »**. On va donc utiliser le capital comme proxy des capitaux propres.

Le ratio de solvabilité permet de mesurer votre capacité à rembourser vos dettes sur le long terme.

```python
data["svblité"]=100*data["CAPITAL"]/data["Dettes fiscales et sociales"]
```

* **taux de pression fiscale**

Indicateur reflétant la charge fiscale effective supportée par l’entreprise, en proportion du résultat ou de la valeur ajoutée

```python
data["tx fiscale"]= 100*data["Impôts sur le résultat"]/data["CHIFFRE D'AFFAIRES (A + B + C + D)"]
```

* **Part de l’impôt sur le résultat rapporté à la valeur ajoutée globale**

Mesure l’intensité fiscale sur l’activité économique créée; utile pour comparer des entreprises ou des secteurs en termes d’efficacité fiscale

```python
data["impot_VA"]=100*data["Impôts sur le résultat"]/data["VALEUR AJOUTEE (XB + RA + RB) + (somme TE à RJ)"]
```

* **Autonomie financière**

Le ratio d'autonomie financière des entreprises, appréhendée par le rapport entre les capitaux propres et le total du bilan. Il mesure la part des capitaux propres dans le financement total. Plus il est élevé, moins l’entreprise dépend de la dette.

```python
data["Autonomie Fin"]=100*data["CAPITAL"]/(data["TOTAL PASSIF CIRCULANT"]+data["TOTAL RESSOURCES STABLES"]).abs()
```

* **Rentabilité financière**

Ratio de rentabilité financière = résultat net / capitaux propres. Elle  exprime la capacité des capitaux investis par les actionnaires et associés (capitaux propres) à dégager du profit.


```python
data["Rentabilité Fin"]=100*data["RESULTAT NET (XG + XH + RQ + RS)"].abs()/data["CAPITAL"]
```
